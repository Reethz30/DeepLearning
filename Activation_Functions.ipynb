```python
#Sigmoid Function
#This function computes the sigmoid of the input x. It maps any real-valued number into the range (0, 1).
#This makes it useful for tasks where you want to predict probabilities or have a bounded output.

import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
  return 1/(1+np.exp(-x))

x=np.linspace(-10,10,50)
p=sigmoid(x)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.plot(x,p)
plt.show()
```

```python
#Derivative Sigmoid Function.
#The derivative of the sigmoid function is used in backpropagation algorithms in neural networks.
#It tells us how the sigmoid output changes with respect to changes in the input, which is crucial for updating weights during training.

def der_sigmoid(x):
  return sigmoid(x)*(1-sigmoid(x))
x=np.linspace(-10,10,50)
p=der_sigmoid(x)
plt.xlabel("x")
plt.ylabel("Der_Sigmoid(x)")
plt.plot(x,p)
plt.show()
```

```python
#Rectified Linear Unit (ReLU) function
#The output of ReLU is in the range [0, ).
#It maps all negative input values to 0 and retains positive values as they are.

def relu(x):
  return np.maximum(0,x)
x=np.linspace(-10,10,50)
p=relu(x)
plt.xlabel("x")
plt.ylabel("RelU(x)")
plt.plot(x,p)
plt.show()
```

```python
#Hyperbolic Tangent Function.
#The output values of the tanh function lie in the interval (-1, 1).

def tanh(x):
  return np.tanh(x)
x=np.linspace(-10,10,50)
p=tanh(x)
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.plot(x,p)
plt.show()
```

```python
#Softmax Function
#It transforms a vector of raw scores (logits) into probabilities that sum up to 1.
#Each element in the output vector is in the range (0, 1) and represents the probability of the corresponding class.

def softmax(x):
  e=np.exp(x)
  return e/np.sum(e)
x=np.linspace(-10,10,50)
p=softmax(x)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.plot(x,p)
plt.show()
```
